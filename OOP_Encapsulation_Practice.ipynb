##Encapsulation - Public variables and functions
####C++

```python
%%writefile OOP-Encapsulation-Examples.cpp
#include<iostream>
using namespace std;
class publicVar
	{
	public:
		int length;
		int breadth;
		int getArea() 
		{           // Function to calculate area
      		return length * breadth;
    	}
};
int main(){
   publicVar obj;
   obj.length = 12;
   obj.breadth = 5;
   cout << "Area = " << obj.getArea();
   return 0;
}
```

```python
%%shell
g++ OOP-Encapsulation-Examples.cpp -o output5
./output5
```

##Encapsulation - Private variables and Public Setter & Getter Functions
####C++

```python
%%writefile OOP-Encapsulation-Examples.cpp
#include<iostream>
using namespace std;
class privateVar
	{
	private:
		int length;
		int breadth;
	public:
		void setLength(int l) 			// Setter function for length
		{       
	      length = l;
	    }
	    void setBreadth(int b) 			// Setter function for breadth
		{
	      breadth = b;
	    }
	    int getLength() 				// Getter function for length
		{         
	      return length;
	    }
	    int getBreadth() 				// Getter function for breadth
		{      
	      return breadth;
	    }
		int getArea() 
		{           // Function to calculate area
      		return length * breadth;
    	}
};
int main(){
   privateVar obj;
   obj.setLength(12);
   obj.setBreadth(5);
   cout << "Length = " << obj.getLength() << endl;
   cout << "Breadth = " << obj.getBreadth() << endl;
   cout << "Area = " << obj.getArea();
   return 0;
}
```

```python
%%shell
g++ OOP-Encapsulation-Examples.cpp -o output5
./output5
```

##Encapsulation - Using Constructors
####C++

```python
%%writefile OOP-Encapsulation-Examples.cpp
#include<iostream>
using namespace std;
class Multiplier
	{
	private:
		int result;
	public:
		Multiplier (int i = 0, int j = 0)		//constructor
		{
			result = i*j;
		}
		int getResult() 
		{           
      		return result;
    	}
};
int main(){
   Multiplier obj(12, 5);
   cout << "Result = " << obj.getResult();
   return 0;
}
```

```python
%%shell
g++ OOP-Encapsulation-Examples.cpp -o output5
./output5
```

##Encapsulation - Classes - Constructor and Public & Private variables
####Python

```python
class Mobile:   
    def __init__(self, name, model, price, color):
        self.__name = name   # __ indicates private variable
        self.__model = model
        self.price = price   # publicly available
        self.color = color
    def getProductInfo(self):
        print('Name: {0}\nModel: {1}\nPrice: Rs. {2}\nColor: {3}\n'.format(self.__name, self.__model, self.price, self.color))
    

Phone = Mobile('OnePlus', 'Nord 2', 29999, 'Gray Sierra')
Phone.getProductInfo()
Phone.__name = 'Redmi'        # will not change variable because its private
Phone.getProductInfo()
```
