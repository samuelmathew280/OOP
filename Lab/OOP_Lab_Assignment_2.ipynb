```python
import math
class Circle:
  def __init__(self, radius):
    self.radius = radius

  def getArea(self):
    return 2*math.pi*self.radius

  def getCircumference(self):
    return (math.pi)*(self.radius)**2

c1 = Circle(5.0)
print("Area:", c1.getArea(), "Circumference:", c1.getCircumference())
```

```python
class Teacher:
  def __init__(self, name, empID, branch, salary):
    self.name = name
    self.empID = empID
    self.branch = branch
    self.salary = salary

  def displayDetails(self):
    print("Name:", self.name + ",", "Employee ID:", str(self.empID) + ",", "Branch:", self.branch + ",", "Salary:", str(self.salary), '\n')

teachers = []
for i in range(0,10):
  name = input("Enter the teacher's name: ")
  empID = int(input("Enter employee ID: "))
  branch = input("Enter branch: ")
  salary = int(input("Enter teacher's salary: "))
  print('\n')
  t = Teacher(name, empID, branch, salary)
  teachers.append(t)

for i in teachers:
  i.displayDetails()
```

```python
class Account:
  def __init__(self, accNo, accName, amount):
    self.accNo = accNo
    self.accName = accName
    self.amount = amount

  def deposit(self):
    amt = int(input("Enter amount to be deposited: "))
    self.amount += amt
    print("Rs. {0} has been deposited.\nCurrent balance: {1}\n".format(amt, self.amount))

  def withdraw(self):
    while(1):
      amt = int(input("Enter amount to be withdrawn: "))
      if self.amount == 0:
        print("No balance in account, can't withdraw any money. Kindly deposit money first.\n")
      elif amt>self.amount:
        print("Insufficient balance. Re-enter money to be withdrawn.\n")
      else:
        self.amount -= amt
        print("Rs. {0} has been withdrawn.\nCurrent balance: {1}\n".format(amt, self.amount))
        break

  def checkBalance(self):
    print("Remaining balance: {0}\n".format(self.amount))

  def displayDetails(self):
    print("Account holder name:", self.accName + ",", "Account number:", str(self.accNo) + ",", "Balance:", self.amount, '\n')

acc = Account(12000032351, "Samuel Mathew", 12000)
acc.displayDetails()
acc.withdraw()
acc.deposit()
acc.checkBalance()
```

```python
class Student:
  def __init__(self, name, USN, marks):
    self.name = name
    self.USN = USN
    self.marks = marks

  def displayDetails(self):
    print("Student name:", self.name + ",", "USN:", self.USN)
    j=0
    for i in self.marks:
      print("Marks in subject {0} = {1}".format(j+1, self.marks[j]))
      j+=1

name = input("Enter student's name: ")
USN = int(input("Enter student's USN: "))
n = int(input("Enter number of subjects appeared for: "))
marks = []
for i in range(0,n):
  mark = int(input("Enter marks for subject {0}: ".format(i+1)))
  marks.append(mark)
std = Student(name, USN, marks)
print('\n')
std.displayDetails()
```

```python
o2Support = 0
generalWard = 0
total = 0
class newPatient:
  def __init__(self):
    self.name = None
    self.dateAdmit = None
    self.symptoms = [] 
    self.o2 = None
    self.dateDischarge = None

  def getInfo(self):
    self.name = input("Enter patient's name: ")
    self.dateAdmit = input("Enter the date of admission: ")
    while(1):
      k = input('Input symptoms, one at a time. Press 1 to escape if finished. ')
      if k == '1':
          if len(self.symptoms) == 0:
              print("At least enter one available blood type.")
              continue
          break
      else:
        self.symptoms.append(k)
    self.o2 = int(input("Enter O2 level (integer): "))
    discharge = input("Has the patient already been discharged? Enter Yes or No: ")
    if discharge.lower().startswith('y'):
      self.dateDischarge = input("Enter the date of discharge: ")
    else:
      if self.o2<90:
        global o2Support
        o2Support +=1
      else:
        global generalWard
        generalWard +=1
    global total
    total +=1
    print()

  def checkSeverity(self):
    if self.dateDischarge != None:
      print("Patient has already been discharged. Hence, patient is assumed to be in good condition.\n")
      return
    if self.o2<90:
      print("Critical! Oxygen level is {0}.\n".format(self.o2))
    else:
      print("Safe. Oxygen level is {0}.\n".format(self.o2))

def noOfPatients():
  print("There are \n{0} patient(s) on oxygen support\n{1} patient(s) in the general ward for observation\nand {2} total patient(s) admitted (including ones that have been discharged).".format(o2Support, generalWard, total))

Patient1 = newPatient()
Patient1.getInfo()
Patient1.checkSeverity()
Patient2 = newPatient()
Patient2.getInfo()
Patient2.checkSeverity()
Patient3 = newPatient()
Patient3.getInfo()
Patient3.checkSeverity()
noOfPatients()
```

```python
class Time:
  def __init__(self, hours = 0, minutes = 0, seconds = 0):
    self.hours = hours
    self.minutes = minutes
    self. seconds = seconds

  def displayTime(self):
    if self.hours > 9:
      print("{0}:".format(self.hours), end='')
    else:
      print("0{0}:".format(self.hours), end='')
    if self.minutes > 9:
      print("{0}:".format(self.minutes), end='')
    else:
      print("0{0}:".format(self.minutes), end='')
    if self.seconds > 9:
      print("{0}".format(self.seconds), end='')
    else:
      print("0{0}".format(self.seconds), end='')

time = Time(2, 34, 9)
time.displayTime()
```

```python
class Student:
  def __init__(self, name, age, roll):
    self.name = name
    self.age = age
    self.roll = roll

Std1 = Student('Rahul', 21, 33)
Std2 = Student ('Rohit', 22, 74)

def compare(Std1, Std2):
  if Std1.age == Std2.age:
    print("Their ages are equal.")
  else:
    print("Their ages are not equal.")

compare(Std1, Std2)
```

```python
class Student:
  institute = "IIIT Dharwad"
  semester = 3
  def __init__(self, name, USN):
    self.name = name
    self.USN = USN

  def printInstance(self):
    print("Name:", self.name + ",", "USN:", self.USN, '\n')
  
  def printClass(self):
    print("Institute name:", self.institute +",", "Semester:", self.semester,'\n')

def explanation():
  print("The class has 2 class variables - institute and semester, 2 instance variables - name and USN, 1 class method - printClass and 1 instance method - printInstance. Instance methods use the instance variables, whereas class methods use the class variables.")

std1 = Student("Rohan", 123654)
std1.printClass()
std1.printInstance()
explanation()
```
