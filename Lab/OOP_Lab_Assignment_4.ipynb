```python
class Node:
    def __init__(self, data, nextNode = None):
        self.value = data
        self.next = nextNode

    def __del__(self):
        print("{0} has been deleted from the stack.".format(self.value))

class Stack(Node):
    def __init__(self, head_val):
        self.head = Node("head")            # head pointer node
        self.size = 0
        self.push(head_val)

    def isEmpty(self):
        if self.size == 0:
            print("Stack is empty.")
        else:
            print("Stack is not empty.")

    def push(self, data):
        node = Node(data, self.head.next)
        self.head.next = node
        self.size += 1
        print("{0} has been added to the stack.".format(data))

    def pop(self):
        if self.size == 0:
            print("Stack is empty!")
            return
        delete = self.head.next
        value = delete.value
        del delete
        self.head.next = self.head.next.next
        self.size -=1
        return value

    def printStack(self):
        if self.size == 0:
            print("Stack is empty!")
            return
        top = self.head.next
        out = ""
        while top:
          out += str(top.value) + " -> "
          top = top.next
        print(out[:-4])

    def __del__(self):
        print("Stack has been deallocated.")

if __name__ == "__main__":
    stack = Stack(1)                # First top node is 1
    for i in range(2, 11):
        stack.push(i)
    stack.printStack()
    for i in range(3, 11):
        stack.pop()
    stack.printStack()
    stack.pop()
    stack.pop()
    stack.printStack()

# Google Colab has an issue where it is printing the statements in the wrong order
# '''Stack has been deallocated.
# head has been deleted from the stack.'''
# is meant to happen at the end
```

```python
class Rectangle1:
    def __init__(self):                                 # Default constructor
        self.length = int(input("Enter length of rectangle: "))
        self.breadth = int(input("Enter breadth of rectangle: "))
    
    def area(self):
        return self.length*self.breadth

class Rectangle2(Rectangle1):
    def __init__(self, length, breadth):                 # Parameterized constructor
        self.length = length
        self.breadth = breadth

rect1 = Rectangle1()
rect2 = Rectangle2(4,7)
print("Area =",rect1.area())
print("Area =",rect2.area())
```

```python
import time
class Node:
    def __init__(self, data, nextNode = None):
        self.value = data
        self.next = nextNode

    def __del__(self):
        print("{0} has been deleted from the queue.".format(self.value))

class Queue(Node):
    def __init__(self, head_val):
        self.rear = Node("rear")            # rear pointer node
        self.head = Node("head")
        self.size = 0
        self.push(head_val)
        self.head.next = self.rear.next            # head pointer node

    def isEmpty(self):
        if self.size == 0:
            print("Queue is empty.")
        else:
            print("Queue is not empty.")

    def push(self, data):
        node = Node(data)
        if self.size == 0:
            self.head.next = self.rear.next = node
        else: 
            self.rear.next.next = self.rear.next = node
        self.size += 1
        print("{0} has been added to the queue.".format(data))

    def pop(self):
        if self.size == 0:
            print("Queue is empty!")
            return
        delete = self.head.next
        value = delete.value
        del delete
        self.head.next = self.head.next.next
        self.size -=1
        return value

    def printStack(self):
        if self.size == 0:
            print("Queue is empty.")
            return
        top = self.head.next
        out = ""
        while top:
          out += str(top.value) + " -> "
          top = top.next
        print(out[:-4])

    def __del__(self):
        print("Queue has been deleted.")

queue = Queue(1)                # First top node is 1
for i in range(2, 11):
    queue.push(i)
queue.printStack()
for i in range(3, 11):
    queue.pop()
queue.printStack()
queue.pop()
queue.printStack()
```
