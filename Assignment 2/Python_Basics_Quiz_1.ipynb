```python
def maxReturn(a,b):
  return max(a,b)

print(maxReturn (1,5))
```

```python
def numberGame(n):
  if n%3 == 0 and n%5 == 0:
    return "numberGame"
  elif n%3 == 0:
    return "number"
  elif n%5 == 0:
    return "Game"
  else:
    return n

print(numberGame(125))
print(numberGame(90))
print(numberGame(66))
print(numberGame(37))
```

```python
demerits = 0
def checkSpeed(speed):
  global demerits
  if speed < 70:
    print("Ok.")
  else:
    dpoints = int((speed-70)/5)
    demerits +=dpoints
    print("Demerit points earned: {0}\nCurrent total: {1}".format(dpoints, demerits))
    if demerits > 12:
      print("License suspended!")
      exit(1)

checkSpeed(64)
checkSpeed(89)
checkSpeed(56)
checkSpeed(104)
checkSpeed(83)
checkSpeed(92)
```

```python
def showNumbers(limit):
  for i in range(limit+1):
    if i%2 == 0:
      print("{0} EVEN".format(i))
    else:
      print("{0} ODD".format(i))

showNumbers(5)
```

```python
def showSigns(rows):
  for i in range(1,rows+1):
    for j in range(i):
      print('&', end='')
    print()

showSigns(5)
```

```python
def checkPrime(n):
  if n == 1:
    prime = False
  else:
    prime = True
  for i in range(2, int(n/2)+1):
    if n%i == 0:
      prime = False
      break
  return prime

def printPrimes(max):
  for i in range(1,max):
    if checkPrime(i) == True:
      print("{0}, ".format(i), end ='')

printPrimes(23)     #Primes BETWEEN 0 and 23
```
