```python
########################
# TYPES OF INHERITANCE #
########################
class Employee:
  def __init__(self, name, ID, gender):
    self.name = name
    self.ID = ID
    self.gender = gender

  def getID(self):
    print("ID:", self.ID,'\n')

# SINGLE INHERITANCE
class Boss(Employee):
  def __init__(self, name, ID, gender, salary):
    super().__init__(name, ID, gender)
  
  def position(self):
    print("Boss")
boss = Boss('Sam', 32456, 'Male', 645542)
boss.position() 
boss.getID()

# HIERARCHIAL INHERITANCE
class Manager(Employee):
  def __init__(self, name, ID, gender, salary):
    super().__init__(name, ID, gender)
  
  def position(self):
    print("Manager")
mgr = Manager('Karen', 75256, 'Female', 645542)
mgr.position() 
mgr.getID()

# MULTILEVEL INHERITANCE
class CO(Boss):
  def __init__(self, name, ID, gender, salary, position):
    super().__init__(name, ID, gender, salary)
  def position(self):
    print("CEO")
ceo = CO('Carmen', 1002, 'Female', 120000, 'CEO')
ceo.position()
ceo.getID()

#MULTIPLE INHERITANCE
class Owner(CO, Boss, Employee):
  def __init__(self, name, ID):
    self.name = name
    self.ID = ID
o = Owner('Sharma', 1001)
o.position()
o.getID()

#HYBRID INHERITANCE
class ManagementHead(Manager, Boss):
  def __init__(self, name, ID):
    self.name = name
    self.ID = ID
Mgmt = ManagementHead('Saakshi', 38692)
Mgmt.position()
Mgmt.getID()
```

### Alternative constructors
* Python allows  to define class methods as well, using the @classmethod decorator and a special first argument cls. 
* The main use of class methods is defining methods that return an instance of the class, but aren't using the same code as __init__().
Example : 
* we are  defining  own class for working with dates, BetterDate. The attributes of the class will be year, month, and day. 
* We want to create another  constructor that creates BetterDate objects given the values for year, month, and day, but you also want to be able to create BetterDate objects from strings like 2020-04-30

```python
# Problem 4: Alternative Constructor 
class MyDate:
    def __init__(self, year, month, day):     # Constructor
      # Python allows multiple variable assignments in one line
      self.year, self.month, self.day = year, month, day    
    # Define a class method from_str
    @classmethod
    def from_str(cls, datestr):
    # Split the string at "-" and  convert each part to integer
        parts = datestr.split("-")
        year, month, day = int(parts[0]), int(parts[1]), int(parts[2])
        return cls(year, month, day)       # Return the class instance  
bd = MyDate.from_str('2021-09-08')   
print(bd.year)
print(bd.month)
print(bd.day)
```

```python
# Alternative Constructor - Practice
class Employee:
  def __init__(self, name, empID, salary):
    self.name = name
    self.empID = empID
    self.salary = salary

  @classmethod
  def intern(cls, emp):
    name = emp[0]
    empID = emp[1]
    salary = emp[2]
    return cls(name, empID, salary)

emp = Employee.intern(["Chris", 414628, '1.5 years'])
print(emp.name)
print(emp.empID)
print(emp.salary)
```

```python
# Define a subclass - Practice
class Vehicle:
  def __init__(self, model, brand, color, price):
    self.model = model
    self.brand = brand
    self.color = color
    self.price = price

  def displayInfo(self):
    print("Model:", self.model+',',"Brand:",self.brand+',',"Color:", self.color+',', "Price:", self.price)

class Bus(Vehicle):
  def __init__(self, model, brand, color, price, capacity):
    super().__init__(model, brand, color, price)
    self.capacity = capacity
  def displayInfo(self):
    Vehicle.displayInfo(self)
    print("Seating capacity:",self.capacity)

bus1 = Bus('Rx-50', 'Mercedes', 'Yellow', 450000, 50)
bus1.displayInfo()
```
